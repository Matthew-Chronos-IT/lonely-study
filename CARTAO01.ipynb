```python
# @title CARTAO 1
nome=str(input("DIGITE SEU NOME:  "))
idade=int(input("DIGITE SUA IDADE:  "))
altura=float(input("DIGITE SUA ALTURA:  "))
cidade=str(input("CIDADE ONDE MORA:  "))
hobby=str(input("QUAL SEU HOBBY:  "))

if idade > 59:
  idd = "Idoso"
elif idade > 17:
    idd = "Adulto"
elif idade > 12:
    idd = "Adolescente"
else:
  idd = "Criança"

print("======= CARTÃO DE VISITA =======")
print("Nome:   ", nome, "\nIdade:  ", idade, " (",idd,")", "\nAltura: ", altura,
      "\nCidade: ", cidade, "\nHobbye: ", hobby)
print("================================")
```
