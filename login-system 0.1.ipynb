```python
#checking if the age is valid:
while True:
  age= int(input("write your age: "))
  if age > 1 :
    break
  else:
    print("write a valid age: ")

print("it is a positive age ")

#categorize age:
while True:
  if age < 18:
    print("You are underage")
  elif age < 65:
    print("You are an adult")
  else:
    print("you are a senior")
  break
```
