```python
# @title LOGIN

# ADD USER INFORMATION
name = input("what is your name ?: ")
key = int(input("choose your password with 4 character: "))
count = int(input("insert your password: "))

# CHECK PASSWORD
while count != key:
  print(f"password incorrect. Try again...")
  count = int(input("insert your password: "))

# VALIDATE ACCESS
while count == key:
  print(f"Access granted. Welcome {name}!")
  count += 1
```

```python
# @title LOGIN 0.2

# ADD USER INFORMATION
name = input("what is your name ?: ").capitalize()
while True:
  key = input("choose your password with 4 character: ")
  if key.isdigit() and len(key) == 4:
    senha = int(key) + 10000
    break
  else:
    print("Password must be 4 digits. Try again.")

# CHECK PASSWORD AND VALIDATE ACCESS
count = int(input("insert your password: "))
count = count + 10000
while count != senha:
  print(f"password incorrect. Try again...")
  count = int(input("insert your password: "))
  count = count + 10000

print(f"Access granted. Welcome {name}!")
```
