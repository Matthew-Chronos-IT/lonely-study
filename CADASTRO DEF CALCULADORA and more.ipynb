```python
# @title CARTAO DE VISITA (def)
def wc ():
  print("SEJA BEM VINDO A SISTEMA DE CARTÕES METALINK TECHNOLOGIA\n")

def info ():
  nm = input("DIGITE SEU NOME: ")
  idd = int(input("DIGITE A SUA IDADE: "))
  prf = input("DIGITE A SUA PROFISSÃO: ")
  dd = [nm, idd, prf]
  return dd

def dados (dd):
  print(f"""===== CARTAO DE VISITA =====
  Nome:        {dd[0]}
  Idade:       {dd[1]}
  Profissão:   {dd[2]}
=========================""")

wc ()
info_user = info ()
dados (info_user)
```

```python
# @title CATEGORIAS USO DE (def)
nome=input("DIGITE SEU NOME:  ")

def idade(idade):
  if idade < 18:
    return "MENOR DE IDADE"
  elif idade < 60:
    return "MAIOR DE IDADE"
  else:
    return "IDOSO"

user = int(input("DIGITE SUA IDADE:  "))
categoria = idade(user)

print("====== USUÁRIO ======")
print("Nomer: ", nome)
print("Idade: ", user)
print("Condição: ", categoria)
print("=====================")
```

```python
# @title CALCULADORA DE MEDIA
nm=input("NOME DO ALUNO:  ")
n1=float(input("DIGITE A PRIMEIRA NOTA:  "))
n2=float(input("DIGITE A SEGUNDA NOTA:  "))
n3= (n1+n2)/2

def result(n3):
  if n3 >= 7:
    return "Aprovado !"
  elif n3 >= 5:
    return "Recuperação"
  else:
    return "reprovado..."

media = result(n3)

print("== RESULTADO ==")
print("Nome: ", nm)
print("Media: ", n3)
print("Status: ", media)
```
